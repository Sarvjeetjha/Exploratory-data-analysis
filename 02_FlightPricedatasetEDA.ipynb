```python
#feature engineering 
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
#drop date of journey column
df.drop('Date_of_Journey',axis=1,inplace=True)
df.info()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Time_hr']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Time_min']=df['Arrival_Time'].str.split(':').str[1]
df['Arrival_Time_hr']=df['Arrival_Time_hr'].astype(int)
df['Arrival_Time_min']=df['Arrival_Time_min'].astype(int)
df.drop('Arrival_Time',inplace=True,axis=1)
df.head()
```

```python
df.info()
```

```python
df['Dep_Time_hr']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_min']=df['Dep_Time'].str.split(':').str[1]
df['Dep_Time_hr']=df['Dep_Time_hr'].astype(int)
df['Dep_Time_min']=df['Dep_Time_min'].astype(int)
df.drop('Dep_Time',inplace=True,axis=1)
```
